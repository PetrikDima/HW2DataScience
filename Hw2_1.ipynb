```python
import pandas as pd
```

```python
tmp = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8", decimal=',', thousands='_')
```

```python
len(tmp)
```

```python
our_table = tmp[10]
our_table
```

Вывести первые строки таблицы с помощью метода head

```python
our_table.head(2)
```

Определите количество строк и столбцов в датафрейме (атрибут shape)

```python
our_table.shape
```

Замените в таблице значения "—" на значения NaN

```python
import numpy as np
new_table_None = our_table.replace('—', np.nan)
new_table_None
```

Определите типы всех столбцов с помощью dataframe.dtypes

```python
new_table_None.dtypes
```

Замените типы не числовых колонок на числовые. Подсказка - это колонки где находился символ "—"


```python
new_table_int = new_table_None.fillna(0)
new_table_int
```

Посчитайте, какая доля пропусков содержится в каждой колонке (используйте методы isnull и sum)

```python
new_table_int.isnull().sum()
```

Удалите из таблицы данные по всей стране, последняя строчка таблицы

```python
new_table_int.drop(27)
```

Замените отсутствующие данные в столбцах средними значениями по этим столбцам (метод fillna)

```python
table_result = new_table_int.astype({'1990': float, '2000': float, '2012': float, '2014': float, '2019': float})
table_result
new_table = new_table_None.fillna(table_result.mean().round(1))
new_table = new_table.astype({'1990': float, '2000': float, '2012': float, '2014': float, '2019': float})
new_table
```

Получите список регионов, где уровень рождаемости в 2019 году был выше среднего по Украине

```python
new_table[new_table['2019'] > new_table['2019'].mean()]
```

В каком регионе была самая высокая рождаемость в 2014 году?

```python
new_table[new_table['2014'] == max(new_table['2014'])]
```

Постройте столбцовую диаграмму рождаемости по регионам в 2019 году

```python
import matplotlib.pyplot as plt
diag = new_table["2019"].plot.bar(rot=27)
```
